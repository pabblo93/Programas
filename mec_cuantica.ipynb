## 1. Principios Fundamentales de la Mecánica Cuántica

### 1.1. Postulados de la Mecánica Cuántica

#### Función de Onda (Ψ)
La función de onda, denotada por Ψ, describe el estado cuántico de una partícula o sistema de partículas.

$$ \Psi(x, t) $$

#### Probabilidad e Interpretación de Born
La probabilidad de encontrar una partícula en una posición determinada está dada por el cuadrado del módulo de la función de onda.

$$ P(x, t) = |\Psi(x, t)|^2 $$

#### Operadores y Observables
Los observables físicos se representan mediante operadores lineales en el espacio de Hilbert. Por ejemplo, el operador de posición y el operador de momento:

$$ \hat{x} \Psi = x \Psi $$
$$ \hat{p} \Psi = -i\hbar \frac{\partial \Psi}{\partial x} $$

### 1.2. Principio de Superposición
Un sistema cuántico puede existir en múltiples estados simultáneamente hasta que se mide.

### 1.3. Principio de Incertidumbre de Heisenberg
No se pueden medir simultáneamente con precisión arbitraria pares de variables conjugadas (por ejemplo, posición y momento).

$$ \Delta x \Delta p \geq \frac{\hbar}{2} $$




## 2. Ecuación de Schrödinger

### 2.1. Ecuación de Schrödinger Independiente del Tiempo
Describe cómo evoluciona la función de onda de una partícula no relativista en un potencial.

$$ -\frac{\hbar^2}{2m} \nabla^2 \Psi + V\Psi = E\Psi $$

### 2.2. Ecuación de Schrödinger Dependiente del Tiempo
Describe la evolución temporal de la función de onda.

$$ i\hbar \frac{\partial \Psi}{\partial t} = \left( -\frac{\hbar^2}{2m} \nabla^2 + V \right) \Psi $$


```python
### 2.1. Ecuación de Schrödinger Independiente del Tiempo
import numpy as np
import matplotlib.pyplot as plt

# Definir constantes
hbar = 1.0545718e-34  # Constante de Planck reducida en J·s
m = 9.10938356e-31    # Masa del electrón en kg
E = 1.60218e-18       # Energía en J (equivalente a 10 eV)

# Definir la función de onda
def psi(x):
    return np.exp(-x**2)

# Definir el potencial (por ejemplo, un pozo de potencial cuadrado)
def V(x):
    return 0.5 * m * (x**2)

# Definir el operador Hamiltoniano
def H(psi, x):
    d2psi_dx2 = np.gradient(np.gradient(psi, x), x)
    return -hbar**2 / (2 * m) * d2psi_dx2 + V(x) * psi

# Definir el dominio del espacio
x = np.linspace(-5, 5, 1000)

# Calcular la función de onda y el Hamiltoniano
psi_x = psi(x)
H_psi = H(psi_x, x)

# Graficar los resultados
plt.plot(x, psi_x, label='Función de Onda $\Psi(x)$')
plt.plot(x, H_psi, label='Hamiltoniano $H\Psi(x)$')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.legend()
plt.title('Ecuación de Schrödinger Independiente del Tiempo')
plt.show()
```
