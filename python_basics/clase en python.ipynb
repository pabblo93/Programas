# Clase
Una clase en Python es un modelo o plantilla que define un tipo de objeto. Puedes pensar en una clase como un plano para construir un edificio. El plano no es un edificio en sí, pero contiene toda la información necesaria para crear uno. De manera similar, una clase define las características (atributos) y comportamientos (métodos) que tendrán los objetos creados a partir de ella.

- Objeto: Un objeto es una instancia de una clase. Cuando creas un objeto a partir de una clase, estás construyendo algo tangible (en el sentido del programa), basado en la plantilla definida por la clase. Los objetos son entidades que tienen:
- Atributos: (o propiedades): Son variables que pertenecen a un objeto y almacenan su estado. Por ejemplo, en una clase Coche, los atributos podrían ser marca, modelo, y año.
- Métodos: Son funciones que pertenecen a un objeto y definen su comportamiento. Por ejemplo, un método en una clase Coche podría ser arrancar(), que simula el arranque del coche.

```python
class Coche:
    def __init__(self, marca, modelo, año):
        self.marca = marca  # Atributo
        self.modelo = modelo  # Atributo
        self.año = año  # Atributo

    def arrancar(self):  # Método
        print(f"El {self.marca} {self.modelo} del año {self.año} está arrancando.")

    def detenerse(self):  # Método
        print(f"El {self.marca} {self.modelo} del año {self.año} se ha detenido.")
```

En este ejemplo:

Coche es una clase que define cómo es un coche en términos de atributos (marca, modelo, año) y comportamientos (arrancar, detenerse).
Atributos: marca, modelo, y año son variables que almacenan el estado de un objeto de tipo Coche.
Métodos: arrancar() y detenerse() son funciones que definen las acciones que un coche puede realizar.
Creando un Objeto
Ahora, podemos crear un objeto (una instancia de la clase Coche):



```python
mi_coche = Coche("Toyota", "Corolla", 2020)
```

Aquí, mi_coche es un objeto de la clase Coche. Tiene sus propios valores para marca, modelo, y año.

Usando los Métodos del Objeto
Podemos interactuar con el objeto utilizando sus métodos:

```python
mi_coche.arrancar()
# Output: El Toyota Corolla del año 2020 está arrancando.

mi_coche.detenerse()
# Output: El Toyota Corolla del año 2020 se ha detenido.
```
